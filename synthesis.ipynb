```python
import pandas as pd
```

```python
df = pd.read_csv("/amazon_com-product_reviews__20200101_20200331_sample.csv")
```

```python
df
```

```python
df["gender"] = ["M" if i%3==0 else "F" for i in range(len(df))]
```

```python
import random
```

```python
random.randint(100,500)
```

```python
df["price"] = [(random.randint(100,500)+1000)/100 for i in range(len(df))]
```

```python
df["price_inr"] = 80*df["price"]
```

```python
df = df.drop(["Crawl Timestamp"], axis = 1)
```

```python
!pip install sdv
```

```python
from sdv.metadata import SingleTableMetadata
```

```python
metadata = SingleTableMetadata()
```

```python
metadata.detect_from_dataframe(data=df)
```

```python
python_dict = metadata.to_dict()
```

```python
 python_dict
```

```python
metadata.update_column(
    column_name='Review Content',
    sdtype="text",
    pii=True
)
```

```python
metadata.update_column(
    column_name='Uniq Id',
    sdtype="id",
    regex_format='[1-9]{3}[A-F]{12}[H-P]{10}[1-9]{3}'
)
```

```python
metadata.update_column(
    column_name = "Url",
    sdtype = "uri",
    pii = True
)
```

```python
metadata.update_column(
    column_name='Review Title',
    sdtype="sentence",
    pii=True
)
```

```python
metadata.update_column(
    column_name='User Id',
    sdtype="id",
    regex_format='A[A-Z]{2}[1-9+A-Z*]{22}[A-Z]{3}'
)
```

```python
from sdv.lite import SingleTablePreset

synthesizer = SingleTablePreset(
    metadata,
    name='FAST_ML'
)
```

```python
synthesizer.fit(df)
```

```python
df = synthesizer.sample(num_rows=100000)
```

```python
df
```

```python
res = df
```

```python
res.to_csv("amazon_synthesized.csv")
```
